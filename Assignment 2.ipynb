```python
plt.figure(figsize=(16,6))
sns.barplot(x='Year', y='Total  Number of Motor Vehicle Crashes', data=df)
plt.title('Year-wise Total Crashes')
plt.xlabel('Year')
plt.ylabel('Total Crashes')
```

### **Inference**
Year-wise Total Number of Motor Vehicle Crashes (Bar plot): The number of motor vehicle crashes fluctuated over the years, with a noticeable peak in 2014.

```python
sns.lineplot(x='Year', y='Number of Fatal Crashes', data=df)
plt.title('Year-wise Fatal Crashes')
plt.xlabel('Year')
plt.ylabel('Fatal Crashes')
```

### **Inference**
The line plot illustrates the trend in fatal crashes over the years, with no clear upward or downward pattern.


```python
sns.boxplot(x='Facility', y='Total  Number of Motor Vehicle Crashes', data=df)
plt.title('Total Crashes Distribution by Facility')
plt.xlabel('Facility')
plt.ylabel('Total Crashes')
plt.xticks(rotation=90, ha='right')
plt.show()
```

### Inference
 The box plot displays the spread and potential outliers in the distribution of total crashes across facilities.


```python
plt.figure(figsize=(16,6))
sns.violinplot(x='Year', y='Total  Number of Motor Vehicle Crashes', data=df)
plt.title('Year-wise Distribution of Total Crashes')
plt.xlabel('Year')
plt.ylabel('Total Crashes')
```

### Inference
 The violin plot reveals the distribution of total crashes across years, with 2014 showing the widest variation.

```python
plt.figure(figsize=(16,6))
sns.pairplot(df[['Year', 'Total  Number of Motor Vehicle Crashes', 'Number of Fatal Crashes']])
plt.show()
```

### Inference
 The pairplot helps visualize relationships and distributions between year, total crashes, and fatal crashes.


```python
sns.heatmap(cor,annot=True)
```

### Inference
 The heatmap displays correlations between numerical variables, with some moderate correlation between year and total crashes.


```python
sns.histplot(df['Total  Number of Motor Vehicle Crashes'], kde=True)
plt.title('Distribution of Total Crashes')
plt.xlabel('Total Crashes')
plt.ylabel('Frequency')
```

### Inference
 The histogram depicts the distribution of total crashes, showing a peak in the lower crash count range.


```python
sns.scatterplot(x='Year', y='Total  Number of Motor Vehicle Crashes', data=df)
plt.title('Scatter Plot: Year vs. Total Crashes')
plt.xlabel('Year')
plt.ylabel('Total Crashes')
```

### Inference
 The scatter plot helps visualize the relationship between year and total crashes, with no clear trend.


```python
labels = ['Facility A', 'Facility B', 'Facility C', 'Facility D']
sizes = [30, 15, 25, 30]
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels,)
plt.title('Distribution of Crashes by Facility')
plt.axis('equal')
```

### Inference
The pie chart displays the proportion of crashes among different facilities.
